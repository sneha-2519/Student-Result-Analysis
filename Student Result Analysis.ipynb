### #Importing Necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("/content/Students data.csv")
df
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.drop("Unnamed: 0", axis = 1)
df
```

```python
plt.figure(figsize = (5,5))
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

### From thge above chart we analysed that, the number of females in the data is more than no.of males


---



```python
gb =  df.groupby("ParentEduc").agg({"MathScore" : "mean", "ReadingScore" : "mean", "WritingScore" : "mean",})
gb
```

```python
sns.heatmap(gb, annot = True)
plt.title("Relationship between parent's education and student's score \n")
plt.show()
```

from above we analysed that, the education of parents have a good impact on their scores

---

```python
gb1 =  df.groupby("ParentMaritalStatus").agg({"MathScore" : "mean", "ReadingScore" : "mean", "WritingScore" : "mean",})
gb1
```

```python
sns.heatmap(gb1, annot = True)
plt.title("Relationship between parent's marital status and student's score \n")
plt.show()
```

from above chart we analysed that, their is no /neglible impact on students score due to their parent's marital status
